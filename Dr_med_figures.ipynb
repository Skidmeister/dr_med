```python
#T-test for weights:
from scipy import stats

def significance_test(test, niveau = 0.05):
    if test.pvalue < niveau:
        print("T Test significant. Null hypothesis can be rejected")
    elif test.pvalue >= niveau:
        print("T Test not significant. Null hypothesis cannot be rejected.")

#t-test between healthy and sick
t_test_1 = ttest_ind(females_not_sick+males_not_sick, females_sick_untreated+females_treated+males_sick_untreated+males_treated)
print(f"T-Test between healthy and sick groups: {t_test_1}")
print(f"n. of samples in group healthy = {len(females_not_sick+males_not_sick)}")
print(f'n. of smaples in sick group = {len(females_sick_untreated+females_treated+males_sick_untreated+males_treated)}')
significance_test(t_test_1, 0.05)
#t-test between sick treated and sick control
t_test_2 = ttest_ind(females_sick_untreated+males_sick_untreated, females_treated+males_treated)
print(f"T-Test between treatment and sick control groups: {t_test_2}")
print(f"n. of samples in sick control group = {len(females_sick_untreated+males_sick_untreated)}")
print(f'n. of smaples in treatment group = {len(females_treated+males_treated)}')
significance_test(t_test_2, 0.05)
```
